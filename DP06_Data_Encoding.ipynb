## Apa itu Data Encoding ?

__Data Encoding__ adalah salah satu tahap praproses data sebelum diproses dengan algoritma <i>machine learning</i>. Dalam mengerjakan projek <i>data science</i> ataupun <i>machine learning</i>, kita akan sangat mungkin menemukan satu atau beberapa fitur yang bertipe kategori, misalnya 'Sangat Baik', 'Baik', 'Tidak Baik'. Nah, komputer tidak dapat memproses data bertipe kategori sehingga kita harus mengubah data tersebut menjadi berbentuk bilangan. Proses ini disebut dengan encoding.

Pertama kita <i>import</i> pandas, kemudian <i>load</i> datasetnya ke dataframe. Kemudian kita tambahkan <i>header</i> yang sesuai untuk menggambarkan masing-masing kolom untuk dapat memudahkan analisis data dan juga memformat <i>missing values</i> __'?'__ menjadi __NaN__ ketika membaca file ke dataframe Pandas menggunakan <code>read_csv()</code>.

```python
import pandas as pd
import numpy as np

column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
              'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
              'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower',
              'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

df = pd.read_csv('../datasets/automobile.data', names=column_names, na_values='?')

df.head()
```

Coba kita cek tipe datanya.

```python
# Mengecek tipe data

df.dtypes
```

<i>Encoding</i> adalah proses yang diterapkan pada data yang bertipe kategori, karena itu kita akan memfilter dataframe <code>df</code> dengan menyingkirkan dahulu kolom bertipe numerik, seperti float dan int, dan hanya menyisakan kolom yang bertipe __object__ agar tidak terlalu banyak kolom yang ditampilkan sehingga memaksimalkan pengamatan proses <i>encoding</i> ini.

Anda dapat memfilter kolom berdasarkan tipe data dengan <code>select_dtypes()</code>.

```python
# Memfilter kolom bertipe object

df_obj = df.select_dtypes(include='object').copy()
df_obj.head()
```

Sebelum proses <i>encoding</i>, kita cek dulu apakah ada <i>missing values</i> pada <code>df_obj</code>.

```python
# Mengecek missing values pada dataframe df_obj

df_obj[df_obj.isnull().any(axis=1)]
```

Ternyata ada dua <i>missing values</i> pada kolom <code>num-of-doors</code>. Kemudian kita isi <i>missing values</i> tersebut dengan __'four'__ yang merupakan nilai yang paling banyak di kolom tersebut.

```python
# Mengecek jumlah data untuk tiap label di kolom num-of-doors

df_obj['num-of-doors'].value_counts()
```

```python
# Mengisi missing values dengan 'four'

df_obj['num-of-doors'].fillna('four', inplace=True)
```

```python
# Mengecek lagi keberadaan missing values

df_obj.isnull().values.any()
```

Sekarang sudah tidak ada <i>missing values</i> di dalam dataframe <code>df_obj</code>. Mari kita berlanjut ke proses <i>encoding</i>.

## Encoding secara manual (Replace)

<i>Encoding</i> berarti mengubah label kategori menjadi bilangan, bukan? Sebelum membahas teknik <i>encoding</i> yang ada di Python, kita akan mencoba teknik <i>replace</i>. Sederhananya, kita akan melakukan <i>replacing</i> secara manual pada kolom bertipe kategori.

Seperti yang kita lihat pada dataframe <code>df_obj</code>, kolom <code>num-of-doors</code> dan <code>num-of-cylinders</code> sejatinya memiliki nilai angka, namun masih ditulis dengan huruf. Nah untuk dua kolom tersebut, misalnya kita ingin mempertahankan nilai tersebut sehingga kita dapat langsung menggantinya dengan <i>method</i> <code>replace()</code>, contohnya __'four'__ kita ganti menjadi __4__.

Sebelumnya melakukan <i>replace</i>, kita harus mengetahui dahulu nilai unik yang ada pada dua kolom tersebut dengan <i>method</i> <code>unique()</code>.

```python
# Menampilkan nilai unik pada kolom num-of-doors

df_obj['num-of-doors'].unique()
```

```python
# Menampilkan nilai unik pada kolom num-of-cylinders

df_obj['num-of-cylinders'].unique()
```

Setelah kita mengetahui nilai unik pada kolom tersebut, kini saatnya kita membuat dictionary untuk mendefinisikan pemetaan perubahannya.

```python
# Membuat dictionary peta perubahan

dic = {'num-of-doors': {'two': 2, 'four': 4},
       'num-of-cylinders': {'four': 4, 'six': 6, 'five': 5, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8}}
```

Lalu kita dapat langsung menggunakan method <code>replace()</code>.

```python
# Menerapkan method replace

df_obj = df_obj.replace(dic)
df_obj.head()
```

Dari output di atas, terlihat bahwa kolom <code>num-of-doors</code> dan <code>num-of-cylinders</code> sudah berubah menjadi bilangan.

## Label Encoding menggunakan .cat.codes

Pendekatan yang kedua untuk proses <i>encoding</i> adalah dengan __Label Encoding__. <i>Label encoding</i> akan mengubah kolom bernilai kategori menjadi bilangan dari 0, 1, 2, dan seterusnya tergantung pada jumlah nilai unik dalam kolom tersebut.

Misalnya dalam kolom <code>body-style</code> ada lima nilai unik, maka perubahannya akan seperti ini:

* __convertible__ : 0
* __hardtop__ : 1
* __hatchback__ : 2
* __sedan__ : 3
* __wagon__ : 4

Urutan angka-angka yang dihasilkan adalah berdasarkan urutan abjad.

Untuk melakukan <i>label encoding</i> di Pandas kita dapat menggunakan <code>.cat.codes</code>. 

Sebelumnya, kita harus ubah tipe data <code>body-style</code> menjadi __category__ terlebih dahulu.

```python
# Mengubah tipe data body-style menjadi category

df_obj['body-style'] = df_obj['body-style'].astype('category')
```

Sekarang coba kita cek tipe data dari kolom <code>body-style</code>.

```python
# Mengecek tipe data kolom body-style

df_obj['body-style'].dtype
```

Setelah mengubah tipe data menjadi 'category', kemudian kita dapat mengaplikasikan <code>.cat.codes</code> dan menyimpan hasilnya ke dalam kolom baru bernama <code>body-style-encode</code>. Menyimpannya di kolom baru adalah salah satu opsi. Anda juga dapat menimpanya langsung pada kolom <code>body-style</code>.

```python
# Mengaplikasikan .cat.codes

df_obj['body-style-encode'] = df_obj['body-style'].cat.codes
df_obj.head()
```

Jika ingin melihat perubahannya, Anda dapat menampilkan hanya dua kolom tersebut secara berdampingan seperti di bawah ini.

```python
# Menampilkan kolom body-style dan body-style-encode

df_obj[['body-style', 'body-style-encode']].head(10)
```

## One-Hot Encoding

__One-Hot Encoding__ adalah teknik yang paling banyak digunakan untuk <i>encode</i> data kategori. Metode ini merepresentasikan data bertipe kategori sebagai vektor biner yang bernilai integer, 0 dan 1. Setiap label kategori akan membentuk kolom baru dengan nama label masing-masing.

<img src="../images/one-hot-encoding.png" width="950" align="center" >

Di Pandas, kita dapat mengaplikasikan one-hot encoding dengan fungsi <code>get_dummies()</code>. Fungsi tersebut dinamakan seperti itu karena menghasilkan variabel indikator/dummy (0 dan 1).

Misalnya kita akan melakukan one-hot encoding untuk kolom <code>drive-wheels</code> yang memiliki tiga label kategori yakni <code>fwd</code>, <code>rwd</code> dan <code>4wd</code>.

```python
# Mengaplikasikan get_dummies()

pd.get_dummies(df_obj, columns=['drive-wheels']).head()
```

Dari output di atas dapat kita lihat ada tiga kolom baru di ujung dataframe yaitu <code>drive-wheels_4wd</code>, <code>drive-wheels_fwd</code>, dan <code>drive-wheels_rwd</code>. Untuk data paling atas misalnya, kolom 4wd dan fwd bernilai 0, sedangkan kolom rwd bernilai 1 karena memang <code>drive-wheels</code>-nya __rwd__.

Jika dirasa nama kolom yang terbentuk terlalu panjang, Anda dapat mengatur <i>prefix</i> kolom tersebut dengan parameter <code>prefix</code> seperti di bawah ini.

```python
# Mengaplikasikan get_dummies() dengan pengaturan prefix

pd.get_dummies(df_obj, columns=['drive-wheels'], prefix=['drive']).head()
```

Anda juga dapat mengaplikasikan <code>get_dummies()</code> untuk lebih dari satu kolom sekaligus seperti di bawah ini.

```python
# Mengaplikasikan get_dummies() untuk dua kolom

pd.get_dummies(df_obj, columns=['drive-wheels', 'fuel-type'], prefix=['drive', 'fuel']).head()
```

## Custom Binary Encoding

Untuk analisis data lebih lanjut, terkadang kita hanya perlu melihat apakah suatu data mengandung label kategori tertentu atau tidak. Jika ini yang Anda butuhkan, Anda dapat mengaturnya dengan menggunakan <code>np.where</code> dari Numpy untuk membuat kondisi.

Sebagai contoh misalnya kita ingin melihat apakah <i>engine-type</i>-nya 'dohc' atau bukan. Kita dapat menulis kodenya seperti di bawah ini.

```python
# contoh custom binary encoding

df_obj['is_dohc'] = np.where(df_obj['engine-type']=='dohc', 1, 0)
```

Kolom baru bernama <code>is_dohc</code> akan terbentuk dan mengandung nilai 1 atau 0, dimana akan bernilai <b>1</b> jika <i>engine-type</i>-nya 'dohc', dan bernilai <b>0</b> jika __bukan__ 'dohc'.

Mari kita <i>print</i> dataframe <code>df_obj</code> untuk melihat hasilnya.

```python
df_obj.head()
```

## Encoding menggunakan library Scikit-Learn

Teknik-teknik sebelumnya yang menggunakan <i>library</i> Pandas memang terlihat lebih <i>simple</i>. Namun jika Anda ingiin membangun model prediktif, teknik <i>encoding</i> menggunakan Scikit-Learn dapat Anda gunakan untuk mengoptimalkan model Anda.

Disini akan dibahas dua teknik yakni dengan <code>OrdinalEncoder</code> dan <code>OneHotEncoder</code> dari Scikit-Learn.

### OrdinalEncoder

Untuk menggunakan <code>OrdinalEncoder</code> di Scikit-Learn, kita harus mengimpornya terlebih dahulu. Kita juga membutuhkan method <code>fit_tranform</code> untuk memproses dan mentransformasi data. 

Misalnya kita ingin melakukan <i>encoding</i> dengan <code>OrdinalEncode</code> pada kolom <code>make</code>.

```python
from sklearn.preprocessing import OrdinalEncoder

ord_encode = OrdinalEncoder()
df_obj['make_encode'] = ord_encode.fit_transform(df_obj[['make']])
```

Sekarang mari kita lihat hasilnya.

```python
df_obj[['make', 'make_encode']].head(20)
```

### OneHotEncoder

Sebelumnya kita telah membahas cara <i>encoding</i> dengan One-Hot di Pandas. Sekarang kita akan mencoba melakukan One-Hot encoding dengan <code>OneHotEncoder</code> dari Scikit-Learn.

Misalnya kita menggunakan kolom <code>body-style</code> untuk percobaan ini.

```python
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder()
encode_result = one_hot.fit_transform(df_obj[['body-style']])

# Mengubah result ke dalam dataframe dan menampilkan beberapa sample datanya
df_onehot = pd.DataFrame(encode_result.toarray(), columns=one_hot.categories_)
df_onehot.sample(10)
```

<code>toarray()</code> digunakan untuk mengkonversi <code>encode_result</code> menjadi format yang dapat dikonversi ke dalam dataframe.

Selajutnya kita gabungkan dengan kolom <code>df_obj</code> yang lainnya menggunakan method <code>join()</code>.

```python
# Menggabungkan dengan df_obj

df_obj = df_obj.join(df_onehot)
```

Mari kita lihat hasilnya.

```python
df_obj.head(10)
```
